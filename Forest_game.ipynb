```python
import random # to generate random numbers

def forest_game():
    health = 100 #player starts with 100 health
    food = 2 #player starts with 2 food
    map_pieces = 1 #player starts with 1 map piece
    print('On a bright sunny day, you wake up in a dense forest. You have no clue how you ended up here. You have 2 food packs.')
    print('You just found a map piece on the floor. You need 10 pieces to escape the forest and reach the city')
    while health >=0 and map_pieces < 10: #if the health of the player is less than 0 or map pieces >= 10, the game ends and the loop stops
        print('---Your game statistics---')
        print('Your health : ' + str(health))
        print('Your food : '+ str(food))
        print('Number of map pieces : ' + str(map_pieces))
        print('What do you do?')
        print('1) Eat food')
        print('2) Go deeper inside the forest')
        print('3) Dig the ground')
        print('4) Harvest the plants near you')
        print('5) Sleep')
        ans = input('Your choice : ')
        while not (ans == '1' or ans == '2' or ans == '3' or ans == '4' or ans == '5'): #to make sure the player is making the correct input
            ans = input('Wrong choice, input again : ')
        #multiple if and elif statements for the choice selected
        if ans == '1':
            if food>0:
                food = food-1
                rnum = random.randrange(1,11) #generates a random integer number between 1 to 10
                if rnum != '10':
                    print('Your health increases by 5')
                    health = health + 5
                else:
                    print('Your food was poisonous, your health decreases by 10')
            else:
                print('You dont have any food')
        elif ans == '2':
            rnum = random.randrange(1,11) #generates a random integer number between 1 to 10
            if rnum == '1' or rnum == '2' or rnum == '3':
                print('YAY, you ran into a banana tree, you have 1 food')
                food = food + 1
            elif rnum == '4' or rnum == '5':
                print('Yay, you found another piece of map')
                map_pieces = map_pieces + 1
            else:
                print('You encountered a snake')
                print('What do you do?')
                print('1) Run away')
                print('2) Fight the snake')
                print('3) Sleep')
                ans = input('Your choice')
                while not (ans == '1' or ans == '2' or ans == '3'):
                    ans = input('Wrong choice, input again : ')
                if ans == '1':
                    rnum = random.randrange(1,11) #generates a random integer number between 1 to 10
                    if rnum == '10':
                        print('The snake bit you. You lose 25 health')
                        health = health - 25
                    else:
                        print('You escaped')
                elif ans == '2':
                    rnum = random.randrange(1,3) #generates a random integer number between 1 to 2
                    if rnum == '1':
                        print('The snake hurt you but you killed it and ate it. You lose 15 health')
                        health = health - 15
                    else:
                        print('You beat the snake and ate it. You gained 15 health')
                        health = health + 15
                else:
                    print('The snake killed you. Your health is 0')
                    health = 0
        elif ans == '3':
            rnum = random.randrange(1,11) #generates a random integer number between 1 to 10
            if rnum == '1' or rnum == '2':
                print('You found 2 pieces of the map')
                map_pieces = map_pieces + 2
            elif rnum == '3' or rnum == '4':
                print('You found a packet of biscuits. Your food increases by 1')
                food = food + 1
            elif rnum == '5':
                print('A poisonous bug in the ground bit you. You lost 15 health')
            else:
                print('You found nothing')
        elif ans == '4':
            rnum = random.randrange(1,6) #generates a random integer number between 1 to 5
            if rnum != '5':
                print('You found a mango tree. Your food increases by 3')
                food = food + 3
            else:
                print('A monkey on the tree punched you. You lost 10 health')
                health  = health - 10
        else:
            rnum = random.randrange(1,3) #generates a random integer number between 1 to 2
            if rnum == '1':
                print('You had a refreshing sleep. Your health increased by 50')
                health = health +50
            else:
                print('A lion ate you while you were sleeping. Your health is 0')
                health = 0
    #loop ends after the game is over
    if health <= 0: #if health is 0 or less than that, you died and lost the game
        print('You died. You lost the game')
    else: #if your health is more than 0 then it means you have found more than or equal to 10 pieces of map. Therefore you won
        print('You found >=10 pieces of the map and escaped the forest. You won')
        
forest_game()
```
